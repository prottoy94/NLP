Html tag remove

```python
import re # Regular expression library for text cleaning
def remove_html_tags(text):
    pattern= re.compile(r'<.*?>') # Regular expression pattern to match HTML tags
    return pattern.sub(r'',text)
```

```python
df['review']=df['review'].apply(remove_html_tags)
df['review']
```

Remove urls

```python
def remove_urls(text):
    pattern= re.compile(r'http\S+') # Regular expression pattern to match URLs
    return pattern.sub(r'',text)
```

```python
text = "Check out this link: http://example.com"
cleaned_text = remove_urls(text)
print(cleaned_text)
```

## Slang Expansion

This section loads the slang dictionary from Excel and expands short forms in a sentence.

```python
from pathlib import Path
import pandas as pd

slang_path = Path("chat_slang_dictionary.xlsx")
slang_df = pd.read_excel(slang_path)

# Use the first two columns as (short_form, full_form) to stay robust to column names.
short_col, full_col = slang_df.columns[:2] # Get the first two columns for short and full forms
slang_map = {
    str(k).strip().upper(): str(v).strip()
    for k, v in zip(slang_df[short_col], slang_df[full_col])
    if pd.notna(k) and pd.notna(v)
}
slang_map
```

```python
def expand_slang(text, mapping):
    words = []
    for token in text.split():
        key = token.upper()
        words.append(mapping.get(key, token))
    return " ".join(words)
```

```python
example = "FYI delhi is the capital of india."
expanded = expand_slang(example, slang_map)
expanded
```

Removing the stop words

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x) # Join the list of words into a single string and return it
```

Remove emojis

```python
import re

def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r"", text)
```

```python
remove_emoji("Loved the movie. It was 😘😘")
```

```python
remove_emoji("Lmao 😂😂")
```

Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations = "?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word)))
```
